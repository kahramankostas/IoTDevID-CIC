```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add
path="./csvs"
files_add=find_the_way(path,'.csv')
files_add
```

```python
import csv
essiz=[]
for i in files_add:
    df=pd.read_csv(i,usecols=["Label"])
    a=df["Label"].unique()
    for j in a:
        if j not in essiz:
            essiz.append(j)
 

for i in files_add:
    df=pd.read_csv(i,usecols=["Label"])
    #print(df.groupby("Label").size())
    a=dict(df.groupby("Label").size())
    for j in essiz:
        if j not in a:
            a[j]=0
    del a[np.nan]
    a["0"]=i
    my_dict=dict(sorted(a.items()))
    name=i.replace("FE.csv","_size.csv")
    with open(name, 'w') as f:  # You will need 'wb' mode in Python 2.x
        w = csv.DictWriter(f, my_dict.keys())
        w.writeheader()
        w.writerow(my_dict)
```

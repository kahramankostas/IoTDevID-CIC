### Discovering Labels

```python
def target_name(name):
    df = pd.read_csv(name,usecols=["Label"])
    target_names=sorted(list(df["Label"].unique()))
    return target_names
```

```python
def folder(f_name): #this function creates a folder.
    try:
        if not os.path.exists(f_name):
            os.makedirs(f_name)
    except OSError:
        print ("Tthe folder could not be created!")
```

```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if ("_confusion matrix" not in file) and ("CB.csv" not in file) :
                    files_add.append(os.path.join(r, file))  
            
    return files_add
```

```python
files_add=find_the_way("./1/results/vs",".csv",)
```

```python
essiz=[]
for i in files_add:
    temp=i.split("@")[1]
    if temp not in essiz:
        essiz.append(temp)
```

```python
len(essiz)
```

```python
results=[['Train',"Test",'Acc', 'b_Acc', 'Prec', 'Rec',
   'F1', 'kap', 'ROC', 'tra-T', 'test-T',"Al-time"]]

for i in essiz:

    for ii in files_add:
        name=ii.split("@")[1]
        if i == name:
            df=pd.read_csv(ii)
            del df["T"]
            del df["CV"]
            temp=df.values
            satir=temp[0][0][:-2].split("-")
            if satir[0]!=satir[1]:
                df=df.mean()
                df=list(df.values)
                temp=[satir[0],satir[1]]
                temp=temp+df
                results.append(temp)
            else:
                print(satir)
results = pd.DataFrame (results[1:], columns = results[0])
```

```python
final=[['Train','Acc', 'b_Acc', 'Prec', 'Rec',
   'F1', 'kap', 'ROC', 'tra-T', 'test-T',"Al-time"]]

for i in results["Train"].unique():
    df=results[results["Train"]==i]
    df=df.mean()
    df=list(df.values)
    temp=[i]
    temp=temp+df
    final.append(temp)
final = pd.DataFrame (final[1:], columns = final[0])
final
```

```python
import os
import pandas as pd
import numpy as np
```

```python
def intjoin(list):
    s = [str(i) for i in list]
    res = int("".join(s))
    return(res)
```

```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add
path="./csvs"
filenames=find_the_way(path,".csv")
filenames
```

#### Rename CSV files
#### formed by adding the prefixes A for "Active" and I for "Idle" to the date of data collection.

```python
for i in filenames:
    temp=i[9:-7]
    temp=temp.replace("_","")
    if "Active" in temp:
        temp=temp.replace("Active","")
        temp=f"A{temp}"
    else:
        temp=temp.replace("Idle","")
        temp=f"I{temp}"
    print(temp)
    os.rename(i,f"./csvs/{temp}.csv")
```

#### a CSV file containing the number of packets per device for the sessions is created

```python
lines=[["File","Date","Type",'AMCREST WiFi Camera',
 'Amazon Alexa Echo Dot',
 'Amazon Alexa Echo Spot',
 'Amazon Alexa Echo Studio',
 'Amazon Plug',
 'Arlo Base Station',
 'Arlo Q Camera',
 'Atomi Coffee Maker',
 'Borun/Sichuan-AI Camera',
 'D-Link Mini Camera',
 'D-Link Water Sensor',
 'Eufy HomeBase 2',
 'Globe Lamp ESP_B1680C',
 'Google Nest Mini',
 'Gosund Plug',
 'Gosund Socket',
 'HeimVision Smart WiFi Camera',
 'HeimVision SmartLife Radio/Lamp',
 'Home Eye Camera',
 'LG Smart TV',
 'Luohe Cam Dog',
 'Nest Indoor Camera',
 'Netatmo Camera',
 'Netatmo Weather Station',
 'Philips Hue Bridge',
 'Ring Base Station',
 'SIMCAM 1S (AMPAKTec)',
 'Smart Board',
 'Sonos One Speaker',
 'Teckin Plug',
 'Yutron Plug',
 'iRobot Roomba']]

for i in filenames:
    print(i)
    df=pd.read_csv(i)
    dist=dict(df.groupby("Label").size())
    
    for d in devices:
        if d not in dist:
            dist[d]=0
    dist=dict(sorted(dist.items()))
    temp=i
    if "I" in temp:
        file_type="IDLE"
    else:file_type="ACTIVE"
    line=[i,"20"+temp[8:-4],file_type]
    line+=dist.values()
    lines.append(line)
results = pd.DataFrame (lines[1:], columns = lines[0])
```

```python
#### Generating session ID numbers that identify which devices the sessions do and do not involve
```

```python
raw_ID= np.where(results!=0, 1, 0)
```

```python
Device_ID={}
for i, ii in enumerate(results["File"]):
    Device_ID[ii]=intjoin(raw_ID[i][3:])
    
```

```python
Device_ID
```

```python
# Grouping of sessions with similar device distribution using session IDs
```

```python
vererse_ID={}
for i in Device_ID:
    if Device_ID[i] not in vererse_ID:
        vererse_ID[Device_ID[i]]=[i]
    else:
        vererse_ID[Device_ID[i]].append(i)
```

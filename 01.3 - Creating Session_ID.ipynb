```python
import os
import pandas as pd
import numpy as np
```

```python
def intjoin(list):
    s = [str(i) for i in list]
    res = int("".join(s))
    return(res)
```

```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add
path="./csvs"
filenames=find_the_way(path,".csv")
filenames
```

```python
for i in filenames:
    temp=i[9:-7]
    temp=temp.replace("_","")
    if "Active" in temp:
        temp=temp.replace("Active","")
        temp=f"A{temp}"
    else:
        temp=temp.replace("Idle","")
        temp=f"I{temp}"
    print(temp)
    os.rename(i,f"./csvs/{temp}.csv")
```

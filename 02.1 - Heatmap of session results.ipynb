```python
def hm(cm,target_names,say):
    graph_name=f"./results/cms/{say}_confusion matrix.pdf"   
    #plt.figure(figsize = (5,4))
    plt.figure(figsize = (len(target_names),int((len(target_names)/5)*4)))
    sns.heatmap(cm,xticklabels=target_names, yticklabels=target_names, annot=True, fmt='g')
    plt.ylabel("TRAINING")
    plt.xlabel("TESTING")
    plt.savefig(graph_name,bbox_inches='tight')#, dpi=400)

    plt.show()
    #print(cm)
    print("\n\n\n") 
    cm=pd.DataFrame(heatmap,columns=target_names,index=target_names)
    cm.to_csv(graph_name.replace(".pdf",".csv"))

```

```python
say=0
dataset="./results/vs/"

for no, i in enumerate(datasets):
    say+=1
    heatmap=[]
    target_names=[]
    for ii in i:
        row=[]
        for iii in i:
            file=f"{dataset}/@{ii[7:-4]}@{iii[7:-4]}.csv"
            #print(file)
            
            df=pd.read_csv(file)
            if ii[7:-4]==iii[7:-4]:
                row.append(np.nan)
            else:
                row.append(round(df["F1"].mean(),3))
        target_names.append(ii[7:-4])
        heatmap.append(row)
    
    hm(heatmap,target_names,say)
```
